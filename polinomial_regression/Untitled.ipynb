```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Position_Salaries.csv')
```

```python
X=data.iloc[:,1:-1].values
Y=data.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
simpl_reg=LinearRegression()
simpl_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polyno_reg=PolynomialFeatures(degree=4)
```

```python
X_poly=polyno_reg.fit_transform(X)
```

```python
reg_2=LinearRegression()
reg_2.fit(X_poly,Y)
```

# Visualizing Both

```python
plt.scatter(X,Y,color="red")
plt.plot(X,simpl_reg.predict(X))
plt.title("SimpleRegrssion")
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,reg_2.predict(X_poly))
plt.title("PolynomialRegressionn")
```
