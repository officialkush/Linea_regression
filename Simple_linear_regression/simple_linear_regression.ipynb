```python
import pandas as pd
import numpy as np 
```

## importing Data

```python
data=pd.read_csv('Salary_Data.csv')
```

## Spliting Data into Independent Variable and Target Variable

```python
X=data.iloc[:,:-1].values
Y=data.iloc[:,-1].values
```

```python
X.shape
```

```python
Y.shape
```

# Now Spliting Data into training and test set 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,
                                               Y,
                                               test_size=0.2,
                                               random_state=0)
```

# Now Creating Simple Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

# Now Visualizing The Model

```python
pred_re=reg.predict(X_test)
```

# Visualizing the Training Data 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,reg.predict(X_train),color='Red')
plt.title("Visualization of training Data")
plt.xlabel("Experience of the employee")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_train,reg.predict(X_train),color='Red')
plt.title("Visualization of training Data")
plt.xlabel("Experience of the employee")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(X_train,reg.predict(X_train),color='Red',linewidth=5)
plt.title("Visualization of training Data")
plt.xlabel("Experience of the employee")
plt.ylabel("Salary")
plt.show()
```
